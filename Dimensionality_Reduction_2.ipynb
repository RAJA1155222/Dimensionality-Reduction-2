Q4. How does the choice of number of principal components impact PCA performance?

More components → more variance retained → better reconstruction of original data, but less dimensionality reduction.

Fewer components → higher compression → may lose information → higher reconstruction error.

Usually, choose enough components to explain 95–99% of variance.

Q5. PCA in feature selection and benefits

PCA can be used for feature extraction, which reduces dimensionality by creating new orthogonal features (principal components).

Benefits:

Reduces noise and redundancy in features

Improves computational efficiency

Mitigates curse of dimensionality

Helps avoid overfitting

Q6. Common applications of PCA in data science / ML

Dimensionality reduction – for large feature sets

Visualization – project high-dimensional data to 2D or 3D

Noise reduction – discard components with low variance

Preprocessing – improve performance of classifiers/regressors

Feature engineering – combine correlated features into principal components

Q7. Relationship between spread and variance in PCA

Spread of the data along a direction = how far points are from the mean along that direction

Variance measures the average squared spread along a direction.

PCA maximizes spread/variance in the new feature axes (principal components).

Q8. How PCA uses spread and variance to identify principal components

Steps:

Center the data (subtract mean)

Compute covariance matrix

Find eigenvectors (directions) and eigenvalues (variance)

Principal components = directions with largest eigenvalues

Intuition: The more spread the data has along a direction, the more information that direction contains.

Q9. Handling high variance in some dimensions and low variance in others

PCA naturally focuses on high-variance directions:

High-variance dimensions dominate the principal components

Low-variance dimensions contribute less and may be discarded

If features are on different scales, feature scaling (standardization) is recommended to prevent large-scale features from dominating PCA.
